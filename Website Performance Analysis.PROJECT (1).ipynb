```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df =pd.read_csv("data-export (1).csv")
df
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.drop(index = 0)
df = df.reset_index(drop = True)
df
```

```python
df.columns =['Channel Group','DateHour','Users','Sessions', 'Engaged sessions',	'Average engagement time per session',	'Engaged sessions per user','Events per session','Engagement rate',	'Event count']
```

```python
df.head()
```

```python
df.info()
```

```python
df['DateHour'] = pd.to_datetime(df['DateHour'], format ='%Y%m%d%H', errors ='coerce')
df
```

```python
df.info()
```

```python
numeric_cols = df.columns.drop(['Channel Group','DateHour'])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
                          
```

```python
df.info()
```

```python
df['Hour'] = df['DateHour'].dt.hour
df
```

```python
df.describe()
```

# sessions and users over time period

```python
df.groupby('DateHour')[['Users','Sessions']].sum()
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize =(10,5))
df.groupby('DateHour')[['Users','Sessions']].sum().plot(ax = plt.gca())
plt.title('Sessions and Users OverTime')
plt.xlabel('Date_hour')
plt.ylabel('Count')
plt.show()
```

# Total users by Channel

```python
plt.figure(figsize =(8,5))
sns.barplot(data = df, x='Channel Group', y='Users', estimator= np.sum, palette= 'magma')
plt.title('Total Users Vs Channel')
plt.xticks(rotation = 45)
plt.show()
```

# Average Engagement time by Channel

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Channel Group', y='Average engagement time per session', palette='viridis')
plt.title('Average Engagement Time by Channels')
plt.xticks(rotation =45)
plt.show()
```

# Engagement rate distribution by channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data = df, x ='Channel Group', y='Engagement rate', palette='coolwarm')
plt.title('Engagement Rate Distribution By Channels')
plt.xticks(rotation =45)
plt.show()
```

# Engaged Vs Non-engaged

```python
df.head()
```

```python
df_sessions = df.groupby('Channel Group')[['Sessions','Engaged sessions']].sum().reset_index()
df_sessions
```

```python
df_sessions['Non_engaged'] = df_sessions['Sessions'] - df_sessions['Engaged sessions']
```

```python
df_sessions_melted = df_sessions.melt( id_vars='Channel Group', value_vars =['Engaged sessions','Non_engaged'])
df_sessions_melted
```

```python
plt.figure(figsize=(8,5))

sns.barplot( data= df_sessions_melted, x='Channel Group', y='value', hue='variable')
plt.title('Engaged & Non-Engaged')
plt.xticks(rotation =45)
plt.show()
```

# traffic by hour and channel

```python
df.head()
```

```python
heatmap_data = df.groupby(['Hour','Channel Group'])['Sessions'].sum().unstack().fillna(0)
heatmap_data
```

```python


plt.figure(figsize =(12,6))

sns.heatmap(heatmap_data, cmap='magma',annot =True, fmt='.0f',linewidths=.5)
plt.title('Traffic by Hour & Channel')
plt.xlabel('Channel Group')
plt.ylabel('Hours in a Day')
plt.show()
```
